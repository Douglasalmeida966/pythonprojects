```python
import pandas as pd

tabela = pd.read_csv("Projeto_ciencia_dados.csv")
display(tabela)
```

```python
tabela["samsung"] = pd.to_numeric(tabela["samsung"], errors="coerce")
tabela["LG"] = pd.to_numeric(tabela["LG"], errors="coerce")
tabela["Apple"] = pd.to_numeric(tabela["Apple"], errors="coerce")
display(tabela.info())



```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(tabela)
plt.show
```

```python
sns.heatmap(tabela.corr(),annot=True, cmap="Wistia")
plt.show()
```
